```python
def TiradaDados(counts):
    
    if len(counts)==5:
        print('escalera')
        moves=0

    if len(counts)==4:
        print('Pareja')
        moves=1

    if len(counts)==3:
        if 3 in counts:
            print('Trio')
            moves=3
        else:
            print('Doble Pareja')
            moves=2

    if len(counts)==2:
        if 4 in counts:
            print('Poker')
            moves=5
        else:
            print('Full')
        moves=4

    if len(counts)==1:

        print('Repoker')
        
        moves=6
    return moves

```

```python
def TiradaDados_noprint(counts):
    
    if len(counts)==5:
        moves=0

    if len(counts)==4:
        moves=1

    if len(counts)==3:
        if 3 in counts:
            moves=3
        else:
            moves=2

    if len(counts)==2:
        if 4 in counts:
            moves=5
        else:
            moves=4

    if len(counts)==1:

        
        moves=6
    return moves
```

```python
times=0
movesss=[]
while times<100000:
    
    n=np.random.randint(1,7, size=5)
    #n=[6,6,6,6,6]

    counts=[('ones', 0), ('twos',0), ('threes',0), ('fours',0), ('fives',0), ('sixes',0)]

    i=1
    for nakero77 in counts:
        for j in n:
             if j==i:
                counts[i-1]=(counts[i-1][0], counts[i-1][1]+1)
        i+=1
    evilCounts=np.array([counts[0][1], counts[1][1], counts[2][1], counts[3][1], counts[4][1], counts[5][1]])

    moves=int()    
    evilEvilCounts=[]
    for i in evilCounts:
        if i!=0:
            evilEvilCounts.append(int(i))

    moves=TiradaDados_noprint(evilEvilCounts)

    if evilCounts[5]>=2:
        evilCount1=evilCounts

        throw=np.random.randint(1,7)

        if throw>1:
            moves=moves+throw

        else:
            j=0
            for i in evilCount1:
                if i==1:
                   evilCount1[j]=0
                   break
                j+=1

            evilCount1[0]=evilCount1[0]+1
            evilEvilCount1=[]

            for i in evilCount1:
                if i!=0:
                    evilEvilCount1.append(int(i))

            moves2=TiradaDados_noprint(evilEvilCount1)

            if moves2>moves+throw:
                moves=moves2
            else: 
                moves=moves+throw

    movesss.append(moves)
    times=times+1

np.mean(movesss)

cum=np.zeros(13)

for i in range(0,13):
    for throw in movesss:
        if throw==i:
            cum[i]=cum[i]+1





print([int(cumin) for cumin in cum])

```
